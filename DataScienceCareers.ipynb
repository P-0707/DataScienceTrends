```python
# figure to show average remote ratio over the years using lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='work_year', y='remote_ratio', estimator='mean')
plt.title("Average Remote Ratio Over the Years")
plt.xlabel("Year")
plt.ylabel("Remote Ratio (%)")
plt.show()
```

<h2>Plot: Average Remote Ratio Over the Years</h2>
<p><strong>Type:</strong> Line Plot</p>
<p><strong>Description:</strong> This line plot displays the average remote work ratio by year, illustrating trends in remote work over time. The shaded area around the line represents the confidence interval, indicating variability in the average remote ratio across years.</p>

<h3>Insights:</h3>
<ul>
    <li>The plot shows a significant decline in the remote work ratio from 2020 to 2024.</li>
    <li>In 2020, the remote work ratio was high, peaking at around 70%, likely due to the shift towards remote work during the COVID-19 pandemic.</li>
    <li>Starting in 2021, the ratio began to decrease steadily, indicating a gradual return to on-site or hybrid work models.</li>
    <li>By 2024, the average remote ratio dropped to below 30%, suggesting that remote work opportunities have diminished significantly.</li>
    <li>This trend may reflect companies encouraging employees to return to office spaces post-pandemic or adopting hybrid models.</li>
</ul>


```python
# figure to show salary trends by experience level over years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='work_year', y='salary_in_usd', hue='experience_level', marker='o')
plt.title("Salary Trends by Experience Level Over Years")
plt.xlabel("Year")
plt.ylabel("Salary in USD")
plt.show()
```

<h2>Plot: Salary Trends by Experience Level Over Years</h2>
<p><strong>Type:</strong> Line Plot with Multiple Lines</p>
<p><strong>Description:</strong> This plot shows the salary trends across different experience levels (EX - Executive, SE - Senior, MI - Mid-level, EN - Entry-level) over the years. Each line represents one experience level, and the shaded area indicates salary variability.</p>

<h3>Insights:</h3>
<ul>
    <li>Executives consistently have the highest average salaries, with values well above other levels, hovering around $300,000.</li>
    <li>Entry-level (EN) and mid-level (MI) roles show a gradual increase in salary over time, reflecting incremental salary growth.</li>
    <li>Senior (SE) roles display a stable yet moderate increase in salaries over the years, reaching an average close to $150,000 by 2024.</li>
    <li>Entry-level salaries remain the lowest but show steady growth, suggesting that starting salaries in the industry are rising.</li>
    <li>This trend indicates companies are investing in higher salaries for experienced roles to retain talent, while entry-level roles remain the lowest-paid.</li>
</ul>


```python
df['job_title'].value_counts()
```

```python
# Calculate the average salary for job titles
avg_salary_by_title = df.groupby('job_title')['salary_in_usd'].mean().nlargest(10).reset_index()
avg_salary_by_title.columns = ['job_title', 'average_salary']  # Rename columns for clarity

# Get the top 10 job titles for visualization
top_job_titles = df['job_title'].value_counts().nlargest(10).reset_index()
top_job_titles.columns = ['job_title', 'count']  # Rename columns for clarity

# Set up the 2x1 grid for subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# First plot: Top 10 Job Titles by Average Salary
sns.barplot(x='job_title', y='average_salary', data=avg_salary_by_title, ax=axes[0], hue='job_title', palette='viridis')
axes[0].set_title("Top 10 Job Titles by Average Salary")
axes[0].tick_params(axis='x', rotation=75)
axes[0].set_xlabel("Job Title")
axes[0].set_ylabel("Average Salary in USD")

# Second plot: Top 10 Job Titles Count
sns.barplot(x='job_title', y='count', data=top_job_titles, ax=axes[1], hue='job_title', palette='viridis')
axes[1].set_title("Top 10 Job Titles Count")
axes[1].tick_params(axis='x', rotation=75)
axes[1].set_xlabel("Job Title")
axes[1].set_ylabel("Count")

# Adjust layout for clear spacing
plt.tight_layout()
plt.show()
```

<h2>Plot: Top 10 Job Titles by Average Salary and Count</h2>
<p><strong>Type:</strong> Bar Plot (Side-by-Side Comparison)</p>
<p><strong>Description:</strong> The first bar plot displays the top 10 job titles with the highest average salaries, while the second plot shows the top 10 job titles by count. The colors reflect the magnitude of the average salary and job count for each role.</p>

<h3>Insights:</h3>
<ul>
    <li><strong>Top Salaries:</strong> Roles like "Analytics Engineering Manager" and "Data Science Team Lead" command the highest salaries, exceeding $350,000. This indicates a premium on specialized and leadership roles in data science and engineering.</li>
    <li><strong>Job Counts:</strong> "Data Engineer" and "Data Scientist" roles have the highest counts, suggesting high demand for these skill sets in the industry.</li>
    <li>Despite higher demand, data engineers and data scientists do not command the highest salaries, likely due to a larger pool of candidates available in these roles compared to specialized or managerial positions.</li>
    <li>Overall, there is a contrast between roles that are highly paid and roles that are widely employed, emphasizing that specialized roles (e.g., "Head of Machine Learning") tend to offer higher compensation but are fewer in number.</li>
    <li>This may suggest a trend where companies invest more in technical leadership roles but continue to hire more standard data engineering and data science roles in larger numbers.</li>
</ul>

```python
# Calculate the required data for each plot
avg_salary_by_company_location = df.groupby('company_location')['salary_in_usd'].mean().nlargest(10)
avg_salary_by_residence = df.groupby('employee_residence')['salary_in_usd'].mean().nlargest(10)
avg_salary_by_currency = df.groupby('salary_currency')['salary_in_usd'].mean()
avg_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()
avg_salary_by_experience.columns = ['experience_level', 'average_salary']

# Set up the 2x2 grid for subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# First plot: Top 10 Average Salaries by Company Location
sns.barplot(x=avg_salary_by_company_location.index, y=avg_salary_by_company_location.values, ax=axes[0, 0])
axes[0, 0].set_title("Top 10 Average Salaries by Company Location")
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].set_xlabel("Company Location")
axes[0, 0].set_ylabel("Average Salary in USD")

# Second plot: Top 10 Average Salaries by Employee Residence
sns.barplot(x=avg_salary_by_residence.index, y=avg_salary_by_residence.values, ax=axes[0, 1])
axes[0, 1].set_title("Top 10 Average Salaries by Employee Residence")
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].set_xlabel("Employee Residence")
axes[0, 1].set_ylabel("Average Salary in USD")

# Third plot: Average Salary by Currency
sns.barplot(x=avg_salary_by_currency.index, y=avg_salary_by_currency.values, ax=axes[1, 0])
axes[1, 0].set_title("Average Salary by Currency")
axes[1, 0].set_xlabel("Currency")
axes[1, 0].set_ylabel("Average Salary in USD")

# Fourth plot: Average Salary by Experience Level
sns.barplot(x='experience_level', y='average_salary', data=avg_salary_by_experience, ax=axes[1, 1], dodge=False)
axes[1, 1].set_title("Average Salary by Experience Level")
axes[1, 1].set_xlabel("Experience Level")
axes[1, 1].set_ylabel("Average Salary in USD")

# Adjust layout for clear spacing
plt.tight_layout()
plt.show()
```
